#**Select**

##1. Descripción del Problema 

En informática, existen diversos problemas a resolver que tienen alguna relación con otras áreas de estudio. El problema de "**Selección**", es uno de estos problemas clásicos, que posee la particularidad de estar altamente relacionado con la estadística, a través del concepto de "**estadístico de orden k°**". Este último término representa al "**k-ésimo**" elemento más pequeño de una *muestra estadística*, y es considerado una herramienta útil en ciertas áreas de la estadística.

El problema de selección debe ser resuelto por un algoritmo que sea capaz de encontrar en el menor tiempo posible un estadístico de orden especificado.

---

#### **Entradas y Salidas:**

**Entrada**: Conjunto de $n$ números diferentes en un arreglo $A$, y un índice $i$ que representa el orden a asignar en el estadístico de orden i°.

**Salida**: El elemento $x$ de $A$, que es mayor que $i-1$ elementos $A$.

##2. Descripción del Algoritmo

Dentro de los algoritmos que le dan una solución al problema de selección, está el "**Select**". Este método *recursivo* toma una secuencia de datos $A$, el índice que indica que dato está buscando la función y un pivote escogido dentro de la misma función, para particionar el arreglo en 2 grupos: Los elementos que son menores que el pivote y los que son mayores. Luego verifica si el pivote escogido es el estadístico de orden buscado, y en caso de que no lo sea, identifica el grupo en donde estaría el elemento a buscar y recurre recursivamente a sí mismo para trabajar en el grupo identificado.

---
###Pasos a Seguir:

A continuación se detallarán los pasos que realiza el algoritmo en su implementación en un arreglo "$A$":

1. La función recibe un arreglo $A$ con $n$ números desordenados y distintos entre sí, además de un índice $i$ que representa que estadístico de orden se está buscando.
2. El algoritmo llama "**Pivot-Selection**", otra función necesaria en la implementación, que se encarga de definir un pivote a partir del arreglo inicial. Este método retorna el pivote a utilizar.
3. Haciendo uso de una función "**Partition**" y el pivote, el algoritmo reorganiza los elementos del arreglo $A$ para que queden separados en 2 grupos: Los números menores al pivote quedarán a su izquierda en el arreglo, mientras que los que sean mayores quedarán a su derecha.
4. Se observa la posición donde quedó el pivote después de "**Partition**", si esta posición representa la posición correcta para el estadístico de orden i°, entonces se encontró la solución.
5. En caso de que no se haya encontrado la solución, se debe encontrar el sub-arreglo que contiene al estadístico de orden. Para esto, se compara la posición $i$ del estadístico de orden con la posición $q$ del pivote, si $i<q$ se asume que elemento a buscar está a la izquierda del pivote, en caso contrario ($i>q$), el elemento está a la derecha del pivote.
6. Se llama recursivamente al algoritmo Select para que opere en el sub-arreglo que tenga el estadístico de orden.

---
###Pivot-Selection:
Para escoger un pivote es posible encontrar varias técnicas, sin embargo, el algoritmo Select ocupa quizá la más adecuada y equilibrada para hallar un buen pivote. La función "**Pivot-Selection**" se encarga de separar el arreglo original en pequeños sub-arreglos, para luego calcular las medianas de cada uno y juntarlas en un arreglo auxiliar. Posterior a aquello, la función obtiene la mediana del arreglo de medianas utilizando llamadas recursivas a sí misma. Cuando el algoritmo obtiene un pivote, lo retorna.

**Pasos de Pivot-Selection**:
1. Divide el arreglo $A$ recibido en $n/5$ conjuntos de $5$ elementos. En caso de que $n$ no sea múltiplo de $5$, entonces el último grupo quedará con menos elementos.
2. Utiliza "**InsertionSort**" para ordenar los elementos de cada conjunto, a medida que los va ordenando obtiene la mediana de cada grupo y agrega esta en un arreglo auxiliar de medianas.
3. Si el arreglo de medianas tiene más de un elemento dentro, se vuelve a llamar a Pivot-Selection utilizando aquel arreglo como entrada.
4. Cuando se obtenga sólo una mediana en el arreglo auxiliar, aquel elemento se retornará como pivote.

---

##3. Implementación del algoritmo

Para la implementación de Select, se utilizará una función partition que deriva del método usado en QuickSort. El código de la función partition es el siguiente:

```python
def insertionSort (A):
    for i in range (1, len(A)):
        #Definir Dato Clave o Pivote.
        key = A[i]
        #Ciclo para Encontrar la Posición del Dato Clave.
        for k in range (i-1, -1, -1):
            #Condición que Verifica si el número Clave está en su posición correcta.
            if A[k] < key: break
            #Intercambiar de posición los números comparados.
            A[k+1] = A[k]
            A[k] = key
    return A #Devolver el Arreglo Ordenado y el Número de Comparaciones realizadas.

def partition (A, piv):
  #Definir límites del arreglo.
  min = 0
  max = len(A)-1
  i = 0
  pivIndex = -1
  for k in range (0, len(A)):
    if (A[k] <= piv):
      A[i], A[k] = A[k], A[i]
      i += 1
      #Encontrar la posición del pivote
      if (A[i-1] == piv):
        pivIndex = i-1
  #Cambiar de posición del pivote.
  A[i-1], A[pivIndex] = A[pivIndex], A[i-1]
  return i-1

def pivotSelection (A):
  #Crear arreglo auxiliar de medianas.
  medians = []
  #Crear arreglo auxiliar de grupos.
  aux = []

  #Dividir el arreglo en conjuntos de 5.
  for i in range (0, len(A)//5):
    #Ordenar un sub-arreglo
    aux = insertionSort(A[5*i:5*i+5])
    #Obtener la mediana del sub-arreglo.
    auxMed = aux[len(aux)//2]
    #Agregar la mediana encontrada al arreglo de medianas.
    medians.append(auxMed)

  #Operar en el conjunto de tamaño menor a 5 si es que existe.
  if ((len(A)//5)*5 != len(A)):
    #Ordenar el sub-arreglo
    aux = insertionSort(A[(len(A)//5)*5:len(A)])
    #Obtener la mediana del sub-arreglo.
    auxMed = aux[(len(aux)//2)]
    #Agregar la mediana encontrada al arreglo de medianas.
    medians.append(auxMed)

  #Verificar si sólo hay una mediana.
  if (len(medians) == 1):
    piv = medians[0]
    return piv

  #Llamar recursivamente a la función
  return pivotSelection(medians)

def Select (A, i):
  #Encontrar un pivote.
  piv = pivotSelection(A)

  #Reorganizar el arreglo con Partition y obtener la posición del pivote.
  pivIndex = partition(A, piv)

  #Verificar si el pivote es el estadístico de orden buscado.
  if (pivIndex == i-1):
    return A[pivIndex]
  if (i-1 < pivIndex):
    return Select(A[0:pivIndex], i)
  if (i-1 > pivIndex):
    return Select(A[pivIndex+1:len(A)], i-pivIndex-1)

  #Retornar el estadístico de orden.
  return A

#Ejemplo:
data = [5, 9, 7, 2, 4, 0, 3, 8, 1, 6]
indBuscado = 6
elemBuscado = Select(data, indBuscado)
print("Elemento Encontrado: ", elemBuscado)
```

```python
def insertionSort (A):
    for i in range (1, len(A)):
        #Definir Dato Clave o Pivote.
        key = A[i]
        #Ciclo para Encontrar la Posición del Dato Clave.
        for k in range (i-1, -1, -1):
            #Condición que Verifica si el número Clave está en su posición correcta.
            if A[k] < key: break
            #Intercambiar de posición los números comparados.
            A[k+1] = A[k]
            A[k] = key
    return A #Devolver el Arreglo Ordenado y el Número de Comparaciones realizadas.

def partition (A, piv):
  #Definir límites del arreglo.
  min = 0
  max = len(A)-1
  i = 0
  pivIndex = -1
  for k in range (0, len(A)):
    if (A[k] <= piv):
      A[i], A[k] = A[k], A[i]
      i += 1
      #Encontrar la posición del pivote
      if (A[i-1] == piv):
        pivIndex = i-1
  #Cambiar de posición del pivote.
  A[i-1], A[pivIndex] = A[pivIndex], A[i-1]
  return i-1

def pivotSelection (A, verbose):
  #Crear arreglo auxiliar de medianas.
  medians = []
  #Crear arreglo auxiliar de grupos.
  aux = []
  i = 0
  
  #Dividir el arreglo en conjuntos de 5.
  if verbose == True:
    print("Cantidad de grupos de 5: ", len(A)//5)
  for i in range (0, len(A)//5):
    #Ordenar un sub-arreglo
    if verbose == True:
      print(f"Grupo {i+1}: ", A[5*i:5*i+5])
    aux = insertionSort(A[5*i:5*i+5])
    if verbose == True:
      print("Ordenar con InsertionSort ---> ", aux)
    #Obtener la mediana del sub-arreglo.
    auxMed = aux[len(aux)//2]
    #Agregar la mediana encontrada al arreglo de medianas.
    medians.append(auxMed)
    if verbose == True:
      print(f"Mediana del grupo {i+1}: ", auxMed)
      print(f"Arreglo de medianas: ", medians)

  #Operar en el conjunto de tamaño menor a 5 si es que existe.
  if ((len(A)//5)*5 != len(A)):
    if verbose == True:
      print("Grupo con menos de 5...")
    #Ordenar el sub-arreglo
    if verbose == True:
      print(f"Grupo {(len(A)//5) + 1}: ", A[(len(A)//5)*5:len(A)])
    aux = insertionSort(A[(len(A)//5)*5:len(A)])
    if verbose == True:
      print("Ordenar con InsertionSort ---> ", aux)
    #Obtener la mediana del sub-arreglo.
    auxMed = aux[(len(aux)//2)]
    #Agregar la mediana encontrada al arreglo de medianas.
    medians.append(auxMed)
    if verbose == True:
      print(f"Mediana del grupo {i+1}: ", auxMed)
      print(f"Arreglo de medianas: ", medians)

  #Verificar si sólo hay una mediana.
  if (len(medians) == 1):
    piv = medians[0]
    if verbose == True:
      print("------------------------------")
      print(f"Se encontró un pivote: {piv}")
      print("--- Sale de pivotSelection ---")
    return piv

  #Llamar recursivamente a la función
  if verbose == True:
    print("------------------------------")
    print("Hay más de una mediana, recurre a PivotSelection")
    print("------------------------------")
  return pivotSelection(medians, True)

def Select (A, i, verbose):
  #Encontrar un pivote.
  if verbose == True:
    print("--- Entra a pivotSelection ---")
  piv = pivotSelection(A, verbose)

  #Reorganizar el arreglo con Partition y obtener la posición del pivote.
  if verbose == True:
    print("--- Entra a partition ---")
    print(f"Arreglo de Entrada: {A}")
  pivIndex = partition(A, piv)
  if verbose == True:
    print(f"Arreglo de Salida: {A}")
    print("Posición del Pivote:", pivIndex)
    print("--- Sale de partition ---")

  #Verificar si el pivote es el estadístico de orden buscado.
  if (pivIndex == i-1):
    if verbose == True:
      print("------------------------------")
      print("Se encontró el elemento")
    return A[pivIndex]
  if (i-1 < pivIndex):
    if verbose == True:
      print("------------------------------")
      print("El elemento está a la izquierda del pivote")
    return Select(A[0:pivIndex], i)
  if (i-1 > pivIndex):
    if verbose == True:
      print("------------------------------")
      print("El elemento está a la derecha del pivote")
    return Select(A[pivIndex+1:len(A)], i-pivIndex-1)

  #Retornar el estadístico de orden.
  return A

#Ejemplo:
data = [5, 9, 7, 2, 4, 0, 3, 8, 1, 6]
indBuscado = 6
print("Arreglo: ", data, "\nEstadístico de Orden a Buscar: ", indBuscado)
elemBuscado = Select(data, indBuscado, verbose = True)
print("Elemento Encontrado: ", elemBuscado)
```

##4. Correctitud de Select.

Para obtener la correctitud de Select, se utilizará el método de inducción matemática. Para esto se deben probar 2 casos.

1. Probar un caso base para $P(n)$ como $P(1)$
2. Probar que si $P(m)$ es verdad para m < n, entonces P(n) también lo es.

Considerando un caso base de $P(1)$:

- Cuando solo hay un elemento en nuestro subarreglo, el i-ésimo elemento más pequeño corresponderá al único elemento presente.

Considerando el caso $P(m)$,  $m < n$:

- Cuando se obtiene el pivote, este se compara con la posición del i-ésimo elemento a buscar, y en el caso de que no correspondan al mismo elemento, el arreglo se particiona en función de continuar la ejecución del algoritmo en un grupo más reducido:
- Si el i-esimo elemento es menor al pivote se continúa por el subarreglo de la izquierda.
- Si el i-ésimo elemento es mayor al pivote se continúa por la derecha. 

Siguiendo esta lógica, se logra minimizar el tamaño de cada subarreglo $m$ con cada llamada recursiva hasta llegar al caso base, cumpliendo la condición de $P(m)$ es verdad para $m < n$

Con lo anteriormente comprobado, se concluir que la correctitud para el algoritmo Select es correcta.


##5. Tiempo de ejecución de Select.

Para obtener la complejidad temporal del algoritmo se realizará un analisis para cada uno de los pasos que sigue el código. Agregado a lo anterior, se establecerá un lema para establecer una conclusión previa de la complejidad.

**Lema:** el tiempo de ejecución del algoritmo Select es de complejidad $O(n)$. 

- ***Paso 1: Pivot-Selection***

Primero el algoritmo *Pivot-Selection* busca la mediana de las medianas, asumiendo que esta función es correcta, entonces se sabe que su definción es $T(n/5)$, debido que realiza un for recorriendo esa cantidad de elementos.

- ***Paso 2: Partition***

El algoritmo *Partition* tiene una complejidad de tiempo $O(n)$, esto se sabe y puede asumir directamente.

- ***Paso 3: Recursión***

Para determinar el tiempo de ejecución de la función recursiva en *Select*, el cual se genera al separar el arreglo en dos casos si no se ha encontrado el k-ésimo dato al finalizar *Partition*, debemos saber que:
    
- Al menos la mitad de las medianas que se encuentran en *Pivot-Selection* son mayores o igual que la mediana verdadera obtenida al finalizar el algoritmo. Esto nos dice que como mínimo la mitad de los grupos de medianas aportan 3 datos mayores que la mediana verdadera, con excepción del último grupo, ya que este tiene a lo más 5 elementos. A partir de lo anterior surge la siguiente proporción.

  Proporción: $[1/2 * n/5] * 3$ => $3n/10$

- Dado que al menos tres elementos son mayores que la mediana verdadera, entonces significa que la función se repite a lo más 7n/10 veces.

  Proporción: $n - 3n/10$ => $7n/10$

Teniendo en cuenta lo anterior, es posible establecer la relación de recurrencia para el algoritmo completo:

  - Ecuación: $T(n) = T(n/5) + T(7n/10) + O(n)$

Aplicando sustitución para probar el lema que el tiempo de ejecución es $O(n)$:

    T(n) <= cn/5 + 7cn/10 + dn

    T(n) <= 9cn/10 + dn => O(n)

En conclusión, el algoritmo Select tiene complejidad temporal promedio es $O(n)$, también se puede apreciar que tiene constantes muy grandes, debido al proceso de escoger la mediana verdadera y las particiones que realiza.



##6. Experimentación.



###Comparación entre R-Select y Select:

Otro experimento a mostrar es la comparación entre **Select** y otro algoritmo que resuelva el problema de Selección. El algoritmo que se utilizará para comparar en realidad ocupa la misma lógica de Select, pero se distingue en el método utilizado para seleccionar un pivote. **R-Select** utiliza un pivote aleatorio, es decir, el algoritmo elige un número al azar dentro del arreglo para que sea el pivote. Teóricamente, Select tiene una peor partición posible de $[3n/10]$ v/s $[7n/10]$, mientras que R-Select puede llegar a tener peores particiones.

El caso general para evaluar el funcionamiento y eficiencia de los métodos, será encontrar el k-ésimo elemento (con k siendo valor aleatorio que esté dentro de los límites del arreglo) en arreglos de entre 250-1500 elementos aleatorios y distintos.

A continuación se presenta el código del R-Select:

```python
import random

def RandomSelection (A):
  #Escoger un número aleatorio del arreglo
  randNum = random.choice(A)
  #Devolver el pivote.
  return randNum

def R_Select (A, i):
  #Encontrar un pivote.
  piv = RandomSelection(A)

  #Reorganizar el arreglo con Partition y obtener la posición del pivote.
  pivIndex = partition(A, piv)

  #Verificar si el pivote es el estadístico de orden buscado.
  if (pivIndex == i-1):
    return A[pivIndex]
  if (i-1 < pivIndex):
    return Select(A[0:pivIndex], i)
  if (i-1 > pivIndex):
    return Select(A[pivIndex+1:len(A)], i-pivIndex-1)

  #Retornar el estadístico de orden.
  return A

#Ejemplo:
data = [5, 9, 7, 2, 4, 0, 3, 8, 1, 6]
indBuscado = 6
elemBuscado = R_Select(data, indBuscado)
print(elemBuscado)
```
